```python
# making series with null values
import pandas as pd
import numpy as np

x=pd.Series([1,2,3,np.nan,5,6,7,np.nan,np.nan,8,9,10])
print(x)
```

```python
# is null ---> returns true if nan else false
print(x.isnull())
```

```python
# total null sum ---. all values that are not null
print(x.dropna())
```

```python
# fillna ---> use to fill nan values by your own logic   fills null value spaces by our own need or logic given

print(x.fillna(x.mean()))

print(x.fillna(x.mode()))

print(x.fillna(x.median()))
```
